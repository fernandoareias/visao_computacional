```python
# Create a function to plot the pixel values on the image
def plot_pixel_values(image, ax):
    height, width = image.shape
    for i in range(height):
        for j in range(width):
            value = image[i, j]
            ax.text(j, i, str(value), ha='center', va='center', color='red', fontsize=14)

# image array
array = [[  0,  51,   0],
         [102, 153, 204],
         [153, 204, 255],
         [255, 255, 255],
         [102, 153, 204],
         [  0,  51,   0]]
image = np.array(array)

# set figure and axes as variables for two plots
fig, axs = plt.subplots(1, 2)

# Plot the image with pixels values
axs[0].imshow(image, cmap='gray')   # cmap = 'gray' to show in gray scale

# to show only entires values in axis
axs[0].set_yticks(np.arange(image.shape[0]))
axs[0].set_xticks(np.arange(image.shape[1]))

# second image
axs[1].imshow(image, cmap='Blues')   # cmap = 'gray' to show in gray scale

plot_pixel_values(image, axs[1])

# to show only entires values in axis
axs[0].set_yticks(np.arange(image.shape[0]))
axs[0].set_xticks(np.arange(image.shape[1]))


# Show the plot
plt.show()
```

# Replace with a patch

```python
from PIL import Image
import numpy as np

# Open the image with Pillow
image = Image.open('Lena_reference_640.jpg')

# Convert the image to a NumPy array
image_array = np.array(image)

# Define the crop coordinates (left, upper, right, lower)
crop_coordinates = (250, 250, 350, 350)  # Example coordinates

# Perform the cropping operation using NumPy array slicing
cropped_image_array = image_array[crop_coordinates[1]:crop_coordinates[3], crop_coordinates[0]:crop_coordinates[2]]

# Define the patch coordinates as a single variable (left, upper, right, lower)
patch_coordinates = (67, 60, 92, 95)

# Define the new position for the patch
new_position = (20, 10)


# Convert the cropped NumPy array back to an image using Pillow
cropped_image = Image.fromarray(cropped_image_array)

# Create a drawing object
draw = ImageDraw.Draw(cropped_image)
# Draw the rectangle on the image
draw.rectangle(patch_coordinates, outline=255, width=1)

# set figure and axes as variables for 1x2 plots
fig, axs = plt.subplots(1, 2, figsize=(12, 8))

# Plot the image with pixels values
axs[0].imshow(cropped_image, cmap='gray')
axs[0].set_title('Original image')


# Extract the patch from the image using NumPy array slicing
patch = cropped_image_array[patch_coordinates[1]:patch_coordinates[3], patch_coordinates[0]:patch_coordinates[2]]


# Paste the patch into the new position in the image array
cropped_image_array[new_position[1]:new_position[1] + patch.shape[0], new_position[0]:new_position[0] + patch.shape[1]] = patch

# Convert the modified NumPy array back to an image using Pillow
modified_image = Image.fromarray(cropped_image_array)

# Display or save the cropped image
axs[1].imshow(modified_image, cmap='gray')
axs[1].set_title('Modified image')

# Show the plot
plt.show()

```

# Continue with the pixel operations

```python
from PIL import Image
import numpy as np

# Open the image with Pillow
image = Image.open('Lena_reference_640.jpg')

# Convert the image to a NumPy array
image_array = np.array(image)

# Define the crop coordinates (left, upper, right, lower)
crop_coordinates = (300, 300, 350, 350)  # Example coordinates

# Perform the cropping operation using NumPy array slicing
cropped_image_array = image_array[crop_coordinates[1]:crop_coordinates[3], crop_coordinates[0]:crop_coordinates[2]]

# set figure and axes as variables for 1x2 plots
fig, axs = plt.subplots(1, 2, figsize=(12, 8))

# Plot the image with pixels values
axs[0].imshow(cropped_image, cmap='gray')
axs[0].set_title('Original image')

# The image is originally stored as uint8 (unsigned integer 8-bit), which means pixel values range from 0 to 255.
print(f"original type: {cropped_image_array.dtype}")

# We are converting the pixel values to float32 to perform operations that could result in values outside the 0-255 range.
# This conversion allows us to manipulate the pixel values using ranges beyond (0-255).
new_image_array = cropped_image_array.astype(np.float32)
print(f"Original image: {new_image_array[0:5,0]}")  # Displaying the first 5 pixel values of the modified image.

# We add 150 to all pixel values, potentially exceeding the range of (0-255).
new_image_array += 150
print(f"Added a value: {new_image_array[0:5,0]}")  # Displaying the first 5 pixel values of the modified image.

# We use np.clip to ensure that pixel values remain within the 0-255 range after addition.
new_image_array = np.clip(new_image_array, 0, 255)
print(f"After clip into (0-255): {new_image_array[0:5,0]}")  # Displaying the first 5 pixel values of the modified image.

# Display the modified image.
# We visualize the modified image with Matplotlib using a grayscale colormap and specifying the range of values.
axs[1].imshow(new_image_array, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Modified image')

# Show the plot
plt.show()
```

# Count the pixel intensities

```python
from PIL import Image
import numpy as np
import matplotlib.animation
import matplotlib.pyplot as plt

# Open the image with Pillow
image = Image.open('Lena_reference_640.jpg')

# Convert the image to a NumPy array
image_array = np.array(image)

# Define the crop coordinates (left, upper, right, lower)
crop_coordinates = (315, 315, 320, 320)  # Example coordinates

# Perform the cropping operation using NumPy array slicing
cropped_image_array = image_array[crop_coordinates[1]:crop_coordinates[3], crop_coordinates[0]:crop_coordinates[2]]

# Convert the cropped NumPy array back to an image using Pillow
cropped_image = Image.fromarray(cropped_image_array)


plt.rcParams["animation.html"] = "jshtml"
plt.rcParams['figure.dpi'] = 75
plt.ioff()
# set figure and axes as variables for 1x2 plots
fig, axs = plt.subplots(1, 2, figsize=(9, 6),gridspec_kw={'width_ratios': [1, 3]})
height, width = cropped_image_array.shape

def animate(t):
    plt.cla()
    # Remove all patches from the plot
    for patch in axs[0].patches:
        patch.remove()

    row = t // width
    column = t % width

    # Plot the image with pixels values
    axs[0].imshow(cropped_image, cmap='gray', vmin=0, vmax=255)
    axs[0].set_title(f'Original image\nPixel value: {cropped_image_array[row, column]}')

    rect = plt.Rectangle((column-0.5, row-0.5), 1, 1, linewidth=3, edgecolor='r', facecolor='none')

    # Agregar el rectngulo al primer subplot
    axs[0].add_patch(rect)

    data = []
    data.extend(cropped_image_array[:row].tolist())
    data.extend(cropped_image_array[row, :column+1].tolist())

    data = [item if isinstance(item, list) else [item] for item in data]
    data = [item for sublist in data for item in sublist]

    axs[1].hist(data, bins=range(0, 256, 10), edgecolor='black')
    axs[1].set_title('Histogram')
    axs[1].set_ylim(0, 10)


matplotlib.animation.FuncAnimation(fig, animate, frames=25)
```

# Histogram with pillow

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Open the image with Pillow
image = Image.open("Lena_reference_640.jpg")

# Convert the image to a NumPy array
image_array = np.array(image)

# Create a figure with subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the image in the first subplot
axs[0].imshow(image_array, cmap='gray')
axs[0].set_title('Image')

# Plot the histogram of the red channel in the second subplot
axs[1].hist(image_array.ravel(), bins=256)
axs[1].set_title('Histogram 256 bins')

# Display the plot
plt.show()
```

```python
from PIL import Image, ImageEnhance

# Open the image
image = Image.open("Lena_reference_640.jpg")

# Increase or decrease the brightness (0.0 is completely black, 1.0 is original brightness, >1.0 increases brightness)
brightness_factor = 1.2  # Adjust this value as needed

# Create an ImageEnhance object with the image
enhancer = ImageEnhance.Brightness(image)

# Adjust the brightness of the image
brightened_image = enhancer.enhance(brightness_factor)

# Convert the image to a NumPy array
image_array = np.array(brightened_image)

# Create a figure with subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the image in the first subplot
axs[0].imshow(image_array, cmap='gray', vmin=0, vmax=255)
axs[0].set_title('Image')

# Plot the histogram of the red channel in the second subplot
axs[1].hist(image_array.ravel(), bins=100)
axs[1].set_title('Histogram 100 bins')
axs[1].set_xlim(0, 256)

# Display the plot
plt.show()
```
