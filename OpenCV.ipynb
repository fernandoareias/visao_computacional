# Thresholding techniques using OpenCV

```python
# organizing imports
import cv2
import numpy as np
import matplotlib.pyplot as plt

# path to input image is specified and
# image is loaded with imread command
image = cv2.imread('coins_Br.jpg')
# Convert from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# cv2.cvtColor is applied over the
# image input with applied parameters
# to convert the image in grayscale
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# applying different thresholding
# techniques on the input image
# all pixels value above 120 will
# be set to 255
_, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
_, thresh2 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)
_, thresh3 = cv2.threshold(img, 120, 255, cv2.THRESH_TRUNC)
_, thresh4 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO)
_, thresh5 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO_INV)



Titles =["Original", 'Binary Threshold', 'Binary Threshold Inverted', 'Truncated Threshold', 'Set to 0', 'Set to 0 Inverted']
images =[image, thresh1, thresh2, thresh3, thresh4, thresh5]

plt.figure(figsize=(12, 6))
plt.subplots_adjust(wspace=0.3)

plt.subplot(2, 3, 1)
plt.title(Titles[0])
plt.imshow(images[0])

for i in range(5):
    plt.subplot(2, 3, i + 2)
    plt.title(Titles[i+1])
    plt.imshow(images[i+1], cmap='gray')

plt.show()
```

# Otsu Thresholding

```python
# organizing imports
import cv2
import numpy as np

# path to input image is specified and image is loaded with imread command
image2 = cv2.imread('coins_Br.jpg')
# Convert from BGR to RGB
image = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

# cv2.cvtColor is applied over the image input with applied parameters to convert the image in grayscale
img = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# applying Otsu thresholding as an extra flag in binary thresholding
ret, thresh1 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# image [y,x,c]

plt.figure(figsize=(20,6))
plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(image2[::-1,:,::-1])

plt.subplot(1,3,2), plt.hist(img.ravel(), 256)
plt.title('Histogram')
plt.axvline(x=ret, color='r', linestyle='dashed', linewidth=2)

plt.subplot(1, 3, 3)
plt.title("Otsu Threshold")
plt.imshow(thresh1, cmap='gray')

plt.show()
```

# Otsu Equation

```python
# Set total number of bins in the histogram
bins_num = 255
thresholds = np.arange(1, bins_num)
sig_c = np.ones_like(thresholds) * np.inf

# Vectorizes image
img_v = img.ravel()

N = len(img_v)

for i, t in enumerate(thresholds):

  bg = img_v[img_v<=t]
  fg = img_v[img_v>t]

  wb = len(bg)
  wf = len(fg)

  sig_b = np.std(bg)
  sig_f = np.std(fg)

  sig_c[i] = wb*sig_b + wf*sig_f

plt.figure(figsize=(10,6))
plt.plot(thresholds, sig_c+1)
plt.title('Thresholds vs  within-cluster variance')
plt.show()

# Minimizes within-cluster variance function val
index_of_min_val = np.argmin(sig_c)

threshold = thresholds[index_of_min_val]
print("Otsu's algorithm implementation thresholding result: ", threshold)

plt.figure(figsize=(20,6))
plt.subplot(1,3,1), plt.imshow(img2,cmap = 'gray')
plt.title('Original Leaf Image' ), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2), plt.hist(img2.ravel(), 256)
plt.axvline(x=threshold, color='r', linestyle='dashed', linewidth=2)
plt.title('Histogram'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3), plt.imshow(imgf,cmap = 'gray')
plt.title('Otsu thresholding'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Set total number of bins in the histogram
bins_num = 256

# Get the image histogram
hist, bin_edges = np.histogram(image, bins=bins_num)

# # Get normalized histogram if it is required
# if is_normalized:
#     hist = np.divide(hist.ravel(), hist.max())

# Calculate centers of bins
bin_mids = (bin_edges[:-1] + bin_edges[1:]) / 2.

# Iterate over all thresholds (indices) and get the probabilities w1(t), w2(t)
weight1 = np.cumsum(hist)
weight2 = np.cumsum(hist[::-1])[::-1]

# Get the class means mu0(t)
mean1 = np.cumsum(hist * bin_mids) / weight1
# Get the class means mu1(t)
mean2 = (np.cumsum((hist * bin_mids)[::-1]) / weight2[::-1])[::-1]

inter_class_variance = weight1[:-1] * weight2[1:] * (mean1[:-1] - mean2[1:]) ** 2

# Maximize the inter_class_variance function val
index_of_max_val = np.argmax(inter_class_variance)

threshold = bin_mids[:-1][index_of_max_val]
print("Otsu's algorithm implementation thresholding result: ", threshold)

plt.figure(figsize=(20,6))
plt.subplot(1,3,1), plt.imshow(img2,cmap = 'gray')
plt.title('Original Leaf Image' ), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2), plt.hist(img2.ravel(), 256)
plt.axvline(x=threshold, color='r', linestyle='dashed', linewidth=2)
plt.title('Histogram'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3), plt.imshow(imgf,cmap = 'gray')
plt.title('Otsu thresholding'), plt.xticks([]), plt.yticks([])
plt.show()
```

# Adaptive Thresholding

```python
# organizing imports
import cv2
import numpy as np
import matplotlib.pyplot as plt

# path to input image is specified and image is loaded with imread command
image = cv2.imread('book_page.jpg')

# cv2.cvtColor is applied over the image input with applied parameters to convert the image in grayscale
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# applying different thresholding techniques on the input image
thresh1 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                          cv2.THRESH_BINARY, 199, 5)

thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 199, 5)

Titles =["Original", 'Adaptive Mean', 'Adaptive Gaussian']
images =[image, thresh1, thresh2]

plt.figure(figsize=(12, 6))
plt.subplots_adjust(wspace=0.3)

plt.subplot(1, 3, 1)
plt.title(Titles[0])
plt.imshow(images[0])

for i in range(2):
    plt.subplot(1, 3, i + 2)
    plt.title(Titles[i+1])
    plt.imshow(images[i+1], cmap='gray')

plt.show()
```
